```python
# Importar las bibliotecas necesarias
from math import pi, cos, sin
from meshpy.tet import MeshInfo, build
from meshpy.geometry import generate_surface_of_revolution, EXT_OPEN, GeometryBuilder
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Definir el radio del objeto
r = 3

# Definir el número de puntos para la generación de la superficie
points = 10
dphi = pi / points

# Función auxiliar para truncar números cercanos a cero
def truncate(r):
    if abs(r) < 1e-10:
        return 0
    else:
        return r

# Generar los puntos de la superficie del objeto
rz = [(truncate(r * sin(i * dphi)), r * cos(i * dphi)) for i in range(points + 1)]

# Crear un objeto GeometryBuilder y agregar la superficie de revolución
geob = GeometryBuilder()
geob.add_geometry(*generate_surface_of_revolution(rz, closure=EXT_OPEN, radial_subdiv=10))

# Crear un objeto MeshInfo y agregar la geometría generada
mesh_info = MeshInfo()
geob.set(mesh_info)

# Generar la malla tetraédrica y guardarla en un archivo vtk
mesh = build(mesh_info)
# mesh.write_vtk("ball.vtk")
```

```python
# Cargar las coordenadas de los nodos y las conexiones entre ellos
vertices = mesh.points
triangles = mesh.elements

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la malla
ax.plot_trisurf(vertices[:, 0], vertices[:, 1], vertices[:, 2], triangles=triangles, cmap=plt.cm.Spectral)

# Personalizar los ejes y la leyenda
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Malla del dominio')
plt.show()
```

```python
import math
import meshpy.triangle as triangle

# Definir el radio del objeto
r = 3

# Definir el número de puntos para la generación de la superficie
num_points = 10
dphi = math.pi / num_points

# Generar los puntos de la superficie del objeto
points = [(r * math.sin(i * dphi), r * math.cos(i * dphi)) for i in range(num_points)]

# Crear la lista de segmentos para definir la restricción de borde
segments = [(i, (i+1)%num_points) for i in range(num_points)]

# Crear un objeto de información de malla
info = triangle.MeshInfo()

# Agregar los puntos y los segmentos a la información de malla
info.set_points(points)
info.set_facets(segments)

# Establecer los parámetros de triangulación
params = triangle.MeshParams()
params.max_volume = 0.1
params.min_angle = 25
params.split_boundary = True
params.refine = True

# Generar la malla
mesh = triangle.build(info, params=params)

# Guardar la malla en un archivo VTK
triangle.write_vtk("ball.vtk", mesh)
```
